Code for On-Axis Light Curve Graphs

```python
import numpy as np
data = np.loadtxt("fnu.txt")
import math
```

```python
import matplotlib.pyplot as plt
from astropy.cosmology import Planck15

t_0 = data[:,0]
F_nu = data[:,1]

theta_0 = math.pi/18
E_52 = 80
n_0 = 1 
z = 1
theta_obs = 2*theta_0
nu_0 = 1*pow(10,14)

t_jet = 6.2*(1+z)*((E_52/n_0)**(1/3))*((theta_0/0.1)**(8/3))

gamma = (theta_0**(-1))*((t_0/t_jet)**(-1/2))   #equation for t_0 > t_jet
a = 1/((1+gamma**2)*(theta_obs**2))             
#t = a*t_0
D = Planck15.luminosity_distance(z=1).cgs.value


a3 = (gamma*theta_obs)**(-6)


F_nu1 = a3*F_nu
t_days = t_0/(86400)       #conversion from seconds to days

L = 4*math.pi*(D**2)*nu_0*F_nu

plt.plot(t_0, F_nu)       #first plot, time in s vs. flux
plt.xlabel('Time (seconds)')
plt.ylabel('Flux density (erg/cm^2/s/Hz)')
plt.xscale('log')
plt.yscale('log')
plt.savefig("Fluxtime_s.png")

```

```python

plt.plot(t_days, F_nu)    #second plot, time in days vs. flux
plt.xlabel('Time (days)')
plt.ylabel('Flux density (erg/cm^2/s/Hz)')
plt.xscale('log')
plt.yscale('log')
plt.savefig("Fluxtime_days.png")
```

```python
plt.plot(t_days, L)      #third plot, time in days vs. luminosity
plt.xlabel('Time (days)')
plt.ylabel('Luminostiy (erg/s)')
plt.xscale('log')
plt.yscale('log')
plt.savefig("Lumtime.png")
```

```python
#Plots of apparent and absolute AB magnitudes

m_AB = -2.5*np.log10(F_nu)-48.60

d = D*(3.24078*pow(10, -19))

M_AB = m_AB-5*np.log10(d)+5


plt.plot(t_days, m_AB)          #plot of time in days vs. apparent magnitude
plt.xlabel('Time (days)')
plt.ylabel('Apparent Magnitude')
plt.xscale('log')
plt.gca().invert_yaxis()
plt.savefig("appmagtime.png")
```

```python

plt.plot(t_days, M_AB)          #plot of time in days vs. absolute magnitude
plt.xlabel('Time (days)')
plt.ylabel('Absolute Magnitude')
plt.xscale('log')
plt.gca().invert_yaxis()
plt.savefig("absmagtime.png")
```

```python
#Function definition to find the Flux Density at any frequency

def F(nu):
    
    Fnu = (nu*F_nu)/nu_0
    return Fnu



plt.plot(t_0, F(1*pow(10,14)), "r-", label='1E14')    #plot for Flux density vs. time at frequency 1E14

plt.plot(t_0, F(1*pow(10,15)), "b-", label='1E15')    #plot for Flux density vs. time at frequency 1E15

plt.plot(t_0, F(1*pow(10,16)), "c-", label='1E16')    #plot for Flux density vs. time at frequency 1E16

plt.xlabel('Time (seconds)')
plt.ylabel('Flux Density (erg/cm^2/s/Hz)')
plt.xscale('log')
plt.yscale('log')
plt.legend()
```

```python
#The light curve for an arbitrary time

def F(t):
    
    Fnu = np.interp(t, t_0, F_nu) 
    return Fnu

print(F(3*pow(10,10)))   #example arbitrary time value



```
